## GLM Einschätzung
- GLM OR **macht Sinn** als Ergänzung zu den Propensity Plots
- Propensity = visuell/deskriptiv (WO brennt es relativ zur WC-Verteilung)
- GLM OR = quantitativ/inferenzstatistisch (WIE STARK ändern sich die Odds pro % WC)
- Zusammen erzählen sie die komplette RQ3-Story
- Empfehlung: GLM OR als Tabelle (platzsparend) ODER als Forest Plot (visuell ansprechend)
- EcoxLC-Stratifizierung gibt dem GLM den entscheidenden Mehrwert vs. Ivey et al. (die nur per Eco hatten)
